### 3190번 - 뱀


* 게임은 NxN 정사각 보드 위에서 진행, 몇몇 칸에는 사과가 놓여져 있음
* 사과를 먹으면 뱀 길이가 늘어난다.
* 시작은 맨 위 맨 좌측 위치 길이는 1 오른쪽 향해 시작
* 매 초마다 이동, 다음과 같은 규칙을 따른다.
* 몸 길이를 늘려 머리를 다음칸에 위치
* 벽, 자기자신과 몸과 부딪히면 게임 끝
* 이동 칸에 사과가 있으면, 사과가 없어지고 꼬리는 움직이지 않음
* 이동 칸에 사과가 없으면, 몸길이를 줄여서 꼬리가 위치한 칸을 비워준다. 즉, 몸길이는 변하지 않는다.

```python
from collections import deque

# 변수, 처음에 오른쪽 이동 dir = 1
n, k, l, y, x, t, ret, idx, dir = 0, 0, 0, 0, 0, 0, 0, 0, 1

# NXN 보드 크기 설정
a = [[0] * 104 for _ in range(104)]
# 방문한 보드 표시
visited = [[0] * 104 for _ in range(104)]

_time = []

# 보드의 크기 N (2<=N<=100), 사과의 개수 K (0<= K <= 100)
n = int(input())
k = int(input())

# 사과의 위치, 첫번째 줄은 정수 행, 두번째 정수는 열 위치 의미, 
# 위치는 모두 다르며 맨위 맨 좌측(1행 1열)에는 사과가 없다.
for _ in range(k):
    y, x = map(int, input().split())
    a[y-1][x-1] = 1

# 뱀의 방향 전환 횟수 l이 주어진다. (1<= L <= 100)
# 뱀의 방향 변환 정보가 주어지는데, 
# 정수 X와 문자 C,D(왼,오)로 이루어져 있으며, 게임 시작 시간으로 X초가 끝난 뒤에 90도 방향 회전
l = int(input())
for _ in range(l):
    t, c = input().split()
    t = int(t)
    if c == 'D':
        _time.append((t, 1))
    else:
        _time.append((t, 3))

time = 0
# dq (0,0), 가변적이므로 덱 사용
dq = deque()
dq.append((0, 0))

# x, y방향으로 이동 벡터
dy = [-1, 0, 1, 0]
dx = [0, 1, 0, -1]

#메인 반복문
while dq:
    time += 1 # 시간을 1 증가
    y, x = dq[0] # 현재 뱀의 머리 위치
    ny = y + dy[dir] # 다음 머리 위치 방향에 따라
    nx = x + dx[dir] # 다음 머리 위치

    # 벽과 충돌 여부 
    if ny >= n or ny < 0 or nx >= n or nx < 0 or visited[ny][nx]:
        break

    # 사과가 없으면, 앞으로 이동만
    if not a[ny][nx]:
        visited[dq[-1][0]][dq[-1][1]] = 0 # 꼬리가 있던 위치의 값 0으로
        dq.pop() # dq에서 마지막 꼬리 요소 제거

    # 사과가 있으면, 길이 늘리면서 전진, pop X
    else:
        a[ny][nx] = 0 #사과 제거
    
    
    visited[ny][nx] = 1 
    dq.appendleft((ny, nx)) # dq 머리 부분 추가
    if idx < len(_time) and time == _time[idx][0]: # 방향전환 명령 확인 시간대 확인
        dir = (dir + _time[idx][1]) % 4 # 방향 확인 후 1은 오른쪽 90도 회전, 3은 왼쪽 90도 회전
        idx += 1

print(time)
```

#### 응용문제 - 로봇청소기

물론이죠. 비슷한 개념을 활용하면서도 약간 다른 요소를 포함하는 문제를 만들어보겠습니다.

"로봇 청소기"

N x M 크기의 직사각형 방이 있습니다. 이 방에는 로봇 청소기가 있으며, 다음과 같은 규칙으로 작동합니다:

1. 로봇 청소기는 현재 위치를 청소합니다.
2. 현재 방향을 기준으로 왼쪽부터 차례대로 탐색합니다:

   a. 왼쪽 방향에 아직 청소하지 않은 공간이 존재한다면, 그 방향으로 회전한 다음 한 칸 전진하고 1번으로 돌아갑니다.


   b. 왼쪽 방향에 청소할 공간이 없다면, 그 방향으로 회전하고 2번으로 돌아갑니다.


   c. 네 방향 모두 청소가 이미 되어있거나 벽인 경우, 바라보는 방향을 유지한 채로 한 칸 후진하고 2번으로 돌아갑니다.


   d. 네 방향 모두 청소가 이미 되어있거나 벽이면서, 뒤쪽 방향이 벽이라 후진도 할 수 없는 경우 작동을 멈춥니다.

3. 방의 각 칸은 벽(1) 또는 빈 공간(0)입니다.
4. 로봇 청소기가 있는 칸은 항상 빈 공간입니다.
5. 로봇 청소기의 초기 위치와 방향이 주어집니다. (0: 북, 1: 동, 2: 남, 3: 서)

입력:
- 첫 줄에 N과 M (3 ≤ N, M ≤ 50)
- 둘째 줄에 로봇 청소기가 있는 칸의 좌표 (r, c)와 방향 d
- 다음 N개의 줄에 방의 상태 (0: 빈 칸, 1: 벽)

출력:
- 로봇 청소기가 청소하는 칸의 개수

이 문제는 뱀 게임과 유사하게 2D 배열에서의 이동과 방향 전환을 다루지만, 청소 로직과 후진 기능 등이 추가되어 더 복잡합니다. 방향 처리, 2D 배열 조작, 조건에 따른 이동 등 비슷한 개념을 활용하면서도 새로운 도전을 제공합니다.

```python
from collections import deque

#방향 벡터
dy = [-1,0,1,0]
dx = [0,1,0,-1]


def clean_room(r,c,d):
    global room, cleaned
    count = 0
    while True:
        #1. 현재 위치 청소
        if room[r][c] == 0:
            room[r][c] = 2 #청소완
            count +=1 

        #2주변 4칸 확인
        cleaned = True
        for _ in range(4):
            nd = (d+3) % 4 
            nr = r + dx[nd]
            nc = c + dy[nd]
            if room[nr][nc] == 0:
                r,c,d = nr,nc,nd
                cleaned = False
                break
            d = nd

        #3. 4방향 모두 청소 혹은 벽인 경우
        if cleaned:
            nr = r - dx[x]
            nc = c - dy[d]
            if room[nr][nc] == 1:
                return count
            r, c = nr, nc #후진
        
#입력 받기
N, M = map(int,input().split())
r,c,d = map(int,input().split())
room = [list(map(int,input().split())) for _ in range(N)]

result = clean_room(r,c,d)
print(result)
```

```python
from copy import deepcopy
from itertools import permutations

def rotate_operation(A, r, c, s):
    #회전 연산을 수행할 배열의 복사본 생성
    B = deepcopy(A)

    #(r-s, c-s)부터 (r+s, c+s)의 정사각형을 시계방향으로 회전
    for layer in range(s):
        #시작점과 끝점 설정
        sy, sx = r-s+layer, c-s+layer
        ey, ex = r+s-layer, c+s-layer

        #현재 레이어의 첫 번쨰 요소를 저장한다.
        temp = B[sy][sx]

        #왼쪽 변 이동
        for i in range(sy,ey):
            B[i][sx] = B[i+1][sx]

        #아래쪽 변 이동
        for j in range(sx, ex):
            B[ey][j] = B[ey][j+1]

        #오른쪽 변 이동
        for i in range(ey,sy,-1): #ey부터 sy까지 1씩 감소시킨다.
            B[i][ex] = B[i-1][ex]
        
        #위쪽 변 이동
        for j in range(ex,sx,-1): 
            B[sy][j] = B[sy][j-1]

        B[sy][sx+1] = temp

    return B


#배열 정보
N, M, K = map(int, input().split())
array = [list(map(int,input().split())) for _ in range(N)]

#회전 연산 정보
operations = []
for _ in range(K):
    r,c,s = map(int,input().split())
    operations.append((r-1, c-1, s))

#모든 가능한 회전 순서에 대해 최소값 계산
min_value = float('inf')
for perm in permutations(range(K)):
    temp_array = deepcopy(array)
    for idx in perm:
        r,c,s = operations[idx]
        temp_array = rotate_operation(temp_array, r,c,s)


    #각 행의 값 중 최소값 계산하기
    min_value = min(min_value, min(sum(row) for row in temp_array))


#최소값 출력
print(min_value)
```

### 모르는 방법들 정리

* permutations란
  * iterable의 모든 가능한 순열(permutation)을 생성한다.
  * for v_idx in permutations(range(k)): 는 0부터 K-1의 숫자로 만들 수 있는 모든 순열을 생성한다.

* min 함수 인자
  * ret = min(ret, solve())
  * solve() 함수는 현재 순열에 따라 회선 연산을 수행, 그 결과의 최소값을 반환한다.
  * min() 함수는 기존의 ret 값과 solve()의 반환값 중 더 작은 값을 선택한다.


* go 함수
  * 현재 위치가 모서리인 경우 방향을 90도 회전(dir - (dir+1)%4)
  * dy, dx 사용해 다음 위치 계산
  * 이미 방문한 위치인 경우 함수를 종료
  * 위치 방문을 표시하고 vv리스트에 추가
  * 재귀적으로 다음 위치로 이동
  

* rotate_all 함수
  * 주어진 중심점 y,x를 기준으로 cnt만큼의 정사각형 테두리를 회전시킨다.
  * 바깥쪽 테두리부터 -> 안쪽으로 회전 수행
  * 각 테두리마다 (sy,sx)와 끝점 (ey, ex)를 설정
  * go를 호출해서 각 회전시킬 요소 위치를 vv에 저장
  * vv의 값들을 추출
  * 값들을 시계방향으로 이동시킨다. values = values[-1:] + values[:-1].
  * 이동된 값들을 원래 배열 b에 다시 할당합니다.
