```python
import sys
from itertools import combinations  # 조합을 생성하기 위한 모듈

# 도시의 크기 n과 남길 치킨집의 수 m을 입력받음
n, m = map(int, input().split())

# 도시의 정보를 2차원 리스트로 입력받음
a = [list(map(int, input().split())) for _ in range(n)]

# 결과값을 저장할 변수, 초기값은 시스템의 최대 정수값
result = sys.maxsize

# 집과 치킨집의 좌표를 저장할 리스트
_home = []
chicken = []

# 도시를 순회하며 집과 치킨집의 좌표를 저장
for i in range(n):
    for j in range(n):
        if a[i][j] == 1:
            _home.append((i, j))  # 집의 좌표 저장
        elif a[i][j] == 2:
            chicken.append((i, j))  # 치킨집의 좌표 저장

# 가능한 모든 치킨집 조합에 대해 반복
for comb in combinations(range(len(chicken)), m):
    ret = 0  # 현재 조합의 치킨 거리 합
    for home in _home:  # 각 집에 대해
        _min = sys.maxsize  # 가장 가까운 치킨집까지의 거리
        for ch in comb:  # 선택된 치킨집들에 대해
            # 맨해튼 거리 계산
            _dist = abs(home[0] - chicken[ch][0]) + abs(home[1] - chicken[ch][1])
            _min = min(_min, _dist)  # 최소 거리 갱신
        ret += _min  # 현재 집의 치킨 거리를 총합에 추가
    result = min(result, ret)  # 전체 치킨 거리의 최소값 갱신

# 최종 결과 출력
print(result)
```

2589번 - 보물섬

```python
from collections import deque  # BFS를 위한 큐 구현에 사용

# 방향 벡터 (상, 우, 하, 좌)
dy = [-1, 0, 1, 0]
dx = [0, 1, 0, -1]

def bfs(y, x):
    global mx  # 전역 변수 mx를 사용
    visited = [[0] * m for _ in range(n)]  # 방문 배열 초기화
    visited[y][x] = 1  # 시작점 방문 표시
    q = deque([(y, x)])  # 큐 초기화

    while q:
        y, x = q.popleft()  # 큐에서 현재 위치 꺼내기
        for i in range(4):  # 4방향 탐색
            ny, nx = y + dy[i], x + dx[i]  # 새로운 위치 계산
            # 범위를 벗어나거나, 이미 방문했거나, 물('W')인 경우 건너뛰기
            if ny < 0 or ny >= n or nx < 0 or nx >= m or visited[ny][nx] or a[ny][nx] == 'W':
                continue
            visited[ny][nx] = visited[y][x] + 1  # 거리 갱신
            q.append((ny, nx))  # 큐에 새로운 위치 추가
            mx = max(mx, visited[ny][nx])  # 최대 거리 갱신

# 입력 받기
n, m = map(int, input().split())
a = [list(input().strip()) for _ in range(n)]

mx = 0  # 최대 거리 초기화

# 모든 육지('L')에 대해 BFS 수행
for i in range(n):
    for j in range(m):
        if a[i][j] == 'L':
            bfs(i, j)

# 결과 출력 (최대 거리 - 1)
print(mx - 1)
```

16234번 - 인구 이동

```python
import sys
sys.setrecursionlimit(10**6)  # 재귀 깊이 제한 늘리기

# 방향 벡터 (상, 우, 하, 좌)
dy = [-1, 0, 1, 0]
dx = [0, 1, 0, -1]

def dfs(y, x):
    global sum, v
    for i in range(4):
        ny, nx = y + dy[i], x + dx[i]
        # 범위를 벗어나거나 이미 방문한 경우 건너뛰기
        if nx < 0 or nx >= n or ny < 0 or ny >= n or visited[ny][nx]:
            continue
        # 인구 차이가 l 이상 r 이하인 경우
        if l <= abs(a[ny][nx] - a[y][x]) <= r:
            visited[ny][nx] = 1
            v.append((ny, nx))
            sum += a[ny][nx]
            dfs(ny, nx)

# 입력 받기
n, l, r = map(int, input().split())
a = [list(map(int, input().split())) for _ in range(n)]

cnt = 0  # 인구 이동 횟수

while True:
    visited = [[0] * n for _ in range(n)]
    flag = False  # 인구 이동 발생 여부
    
    for i in range(n):
        for j in range(n):
            if not visited[i][j]:
                v = [(i, j)]
                visited[i][j] = 1
                sum = a[i][j]
                dfs(i, j)
                
                if len(v) > 1:  # 연합이 형성된 경우
                    new_population = sum // len(v)
                    for y, x in v:
                        a[y][x] = new_population
                    flag = True
    
    if not flag:  # 더 이상 인구 이동이 없으면 종료
        break
    
    cnt += 1

print(cnt)
```

4179번 - 불!

```python
from collections import deque
import sys

input = sys.stdin.readline
INF = int(1e9)

# 방향 벡터 (상, 좌, 하, 우)
dx = [-1, 0, 1, 0]
dy = [0, -1, 0, 1]

def in_range(a, b):
    return 0 <= a < n and 0 <= b < m

# 입력 받기
n, m = map(int, input().split())
a = [list(input().strip()) for _ in range(n)]

# 불과 사람의 이동 시간을 저장할 배열
fire_check = [[INF] * m for _ in range(n)]
person_check = [[0] * m for _ in range(n)]

q = deque()
sx, sy = 0, 0

# 초기 상태 설정
for i in range(n):
    for j in range(m):
        if a[i][j] == 'F':
            fire_check[i][j] = 1
            q.append((i, j))
        elif a[i][j] == 'J':
            sy, sx = i, j

# 불의 확산 BFS
while q:
    y, x = q.popleft()
    for i in range(4):
        ny, nx = y + dy[i], x + dx[i]
        if not in_range(ny, nx):
            continue
        if fire_check[ny][nx] != INF or a[ny][nx] == '#':
            continue
        fire_check[ny][nx] = fire_check[y][x] + 1
        q.append((ny, nx))

# 사람의 이동 BFS
person_check[sy][sx] = 1
q.append((sy, sx))
ret = 0

while q:
    y, x = q.popleft()
    if x == m - 1 or y == n - 1 or x == 0 or y == 0:
        ret = person_check[y][x]
        break
    for i in range(4):
        ny, nx = y + dy[i], x + dx[i]
        if not in_range(ny, nx):
            continue
        if person_check[ny][nx] or a[ny][nx] == '#':
            continue
        if fire_check[ny][nx] <= person_check[y][x] + 1:
            continue
        person_check[ny][nx] = person_check[y][x] + 1
        q.append((ny, nx))

# 결과 출력
if ret != 0:
    print(ret)
else:
    print("IMPOSSIBLE")

# 예제 입력:
# 3 3
# ...
# .J.
# ...
```

12869번 - 뮤탈리스크

```python
from collections import deque

INF = 987654321

# 각 공격의 데미지 조합
_a = [
    [9, 3, 1],
    [9, 1, 3],
    [3, 1, 9],
    [3, 9, 1],
    [1, 3, 9],
    [1, 9, 3]
]

def solve(a, b, c):
    # 3차원 방문 배열 초기화
    visited = [[[0 for _ in range(64)] for _ in range(64)] for _ in range(64)]
    
    q = deque()
    visited[a][b][c] = 1
    q.append((a, b, c))
    
    while q:
        a, b, c = q.popleft()
        if visited[0][0][0]:
            break
        
        for damage in _a:
            na = max(0, a - damage[0])
            nb = max(0, b - damage[1])
            nc = max(0, c - damage[2])
            
            if visited[na][nb][nc]:
                continue
            
            visited[na][nb][nc] = visited[a][b][c] + 1
            q.append((na, nb, nc))
    
    return visited[0][0][0] - 1

# 입력 받기
n = int(input())
a = list(map(int, input().split()))

# 결과 출력
print(solve(a[0], a[1], a[2]))
```

16637번 - 괄호 추가하기

```python
# 숫자와 연산자를 저장할 리스트
num = []
oper_str = []

# 결과값 초기화
ret = -987654321

def oper(a, b, c):
    """
    연산자에 따른 계산을 수행하는 함수
    """
    if a == '+': return b + c
    if a == '-': return b - c
    if a == '*': return b * c

def go(here, _num):
    """
    재귀적으로 모든 가능한 계산 순서를 탐색하는 함수
    """
    global ret
    if here == len(num) - 1:
        # 마지막 숫자에 도달하면 최대값 갱신
        ret = max(ret, _num)
        return
    
    # 다음 숫자와 바로 연산
    go(here + 1, oper(oper_str[here], _num, num[here + 1]))
    
    # 다다음 숫자까지 있다면, 다음 숫자와 다다음 숫자를 먼저 계산
    if here + 2 <= len(num) - 1:
        temp = oper(oper_str[here + 1], num[here + 1], num[here + 2])
        go(here + 2, oper(oper_str[here], _num, temp))

# 입력 받기
n = int(input())
s = input()

# 입력 문자열에서 숫자와 연산자 분리
for i in range(n):
    if i % 2 == 0:
        num.append(int(s[i]))
    else:
        oper_str.append(s[i])

# 계산 시작
go(0, num[0])

# 결과 출력
print(ret)
```
